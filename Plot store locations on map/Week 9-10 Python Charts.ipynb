```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Loading data into dataframe.
costco_df = pd.read_csv('ex5-2/costcos-geocoded.csv')

# Viewing the dataframe.
costco_df.head()
```

```python
# Loading data into dataframe.
ppg_df = pd.read_csv('ex5-2/ppg2008.csv')

# Viewing the dataframe
ppg_df.head()
```

```python
# Plotting heatmap.
sns.set(rc = {'figure.figsize':(25,25)})
sns.set(font_scale = 1.5)
sns.heatmap(ppg_df.iloc[:,1:], yticklabels = ppg_df['Name  '], linewidths = 0.30, annot = True).set(title='NBA Performance - 2008')
```

```python
from mpl_toolkits.basemap import Basemap
```

```python
# Plotting Spatial Chart.
fig = plt.figure(figsize=(7, 7))
m = Basemap(projection='lcc', resolution=None,
            width=7E6, height=7E6, 
            lat_0=45, lon_0=-100,)
m.etopo(scale=0.5, alpha=0.5)

for index, row in costco_df.iterrows():
    Latitude = row['Latitude']
    Longitude = row['Longitude']
    x, y = m(Longitude, Latitude)
    plt.plot(x, y, 'ok')

plt.title('Costco Locations in USA')
```

```python
import plotly.express as px
```

```python
# Plotting Contour Chart.
fig = px.density_contour(ppg_df, x='G', y='PTS', title='2008 NBA - G vs PTS')
fig.update_traces(contours_coloring='fill', contours_showlabels = True)
fig.show()
```
